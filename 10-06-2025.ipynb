## line plot

```python

import matplotlib.pyplot as plt

#sample data
days = list(range(1,32))
temperatures = [68,70,72,75,77,80,82,81,83,81,78,75,72,71,70,71,
                74,77,79,80,82,84,86,88,87,85,82,80,77,75,73]
#create the line plot
plt.plot(days, temperatures, marker='o')
plt.xlabel("day of the month")
plt.ylabel("temperature(f)")
plt.title("daily temperature over a month")
plt.show()
```

## bar plot

```python
import matplotlib.pyplot as plt
categories=['rent','Groceries','Utilities','Entertainment','Transportation']
expenses=[5000,12000,7000,2000,3000]
plt.bar(categories ,expenses)
plt.xlabel("daily expenses categories")
plt.ylabel("cost of expenses")
plt.title("daily necessaries")
plt.show()
```

## scatter plot

```python
study_hours = [2,3,4,5,6,7,8,20,10,50,60,70]
exam_scores = [65,75,60,80,70,75,85,70,92,100,90,60]
plt.scatter(study_hours,exam_scores,label='student Data')
plt.xlabel('study hours')
plt.ylabel('exam scores')
plt.title('study hours vs exam scores')

plt.show()
```

```python
square_footage = [1200,1400,1600,1800,2000,2200,2400,2600,2800,3000]
selling_prices = [250,290,315,380,410,450,500,525,570,610]
plt.scatter(square_footage,selling_prices )
plt.xlabel("square footage")
plt.ylabel("selling price")
plt.title("relationship between the size of houses")

plt.show()
```

## pie chart

```python
categories=['rent','food','Transportation','Entertainment','savings']
values=[35,25,15,10,15]
plt.pie(values,labels = categories,autopct='%1.1f%%')
plt.title("daily expenses")
plt.show()
```
