## introduction to sea born

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
df = sns.load_dataset("tips")
df.head() # this is used to print top most 5 rows
```

```python
# scatter plot
sns.scatterplot(data=df,x="total_bill",y="tip")
```

```python
sns.scatterplot(data=df, x="size", y="tip")
```

```python
df.head(1)
```

```python
sns.scatterplot(data=df, x="total_bill", y="tip", hue="smoker")
```

```python
ax = sns.scatterplot(data = df , x = "total_bill", y = "tip")

ax.set_title("scatterplot total_bill vs tip")
ax.set_xlabel("total bill")
ax.set_ylabel("Tip paied by the customer")
```
