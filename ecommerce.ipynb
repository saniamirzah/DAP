## Load the Data

```python
import pandas as pd
df = pd.read_csv("ecommerce_sales_data.csv")
df
```

## Check Data info and Types

```python
df.info()
```

## Handle Missing Values

```python
df.isnull().sum()
```

```python
# combining year and month into  proper data column
```

```python
df['date']= pd.to_datetime(df['Month'] + ' ' + df['Year'].astype(str), format = '%B %Y')

#Combines the month and year columns, e.g., "March 2024"
```

```python
df['Category'] = df['Category'].str.strip().str.title()
df['Region'] = df['Region'].str.strip().str.title()
#str.strip() → Removes spaces from string values.
#str.title() → Converts each word to title case (first letter uppercase, rest lowercase).
```

## Create New Insightful Columns

### profit margin

```python
df['Profit_Margin'] = (df['Profit']/df['Revenue'])*100

#Calculates profit margin as a percentage:
#Adds the result to a new column called 'Profit_Margin'.
```

#### Revenue per unit

```python
df ['Revenue_Per_Unit'] = df['Revenue']/ df['Units_Sold']

#This creates a new column 'Revenue_Per_Unit' by dividing total revenue by units sold for each row
```

#### Ad Spend to Revenue Ratio

```python
df['Ad_to_Revenue'] = df['Ad_Spend']/ df['Revenue']

# This creates a new column 'Ad_to_Revenue' by dividing Ad_Spend by Revenue, which shows the proportion of revenue spent on advertising for each row
```

```python
df  = df.sort_values('date')

# this sorts the DataFrame in ascending order based on the 'date' column.
```

#### save cleaned file

```python
df.to_csv("cleaned_ecommerce_sales.csv",index=False)
```

```python
#total Revenue
total_revenue = df['Revenue'].sum()
print(f"Total Revenue: ₹{total_revenue:,.2f}")

#df['Revenue'].sum() → Sums up all values in the 'Revenue' column.

#₹{total_revenue:,.2f} → Formats the number with:

#, → Thousand separators

#.2f → Two decimal places

#₹ → Rupee symbol in front

```

```python
#total Profit
total_profit = df['Profit'].sum()
print(f"Total Profit: ₹{total_profit:,.2f}")
```

```python
#Total units sold
total_units = df['Units_Sold'].sum()
print(f"Total Units Sold: {total_units}")
```

```python
# Average customer Rating
avg_rating = df['Customer_Rating'].mean()
print(f"Average Customer Rating: {avg_rating:.2f}")
```

```python
# Average Profit Margin
avg_margin = df['Profit_Margin'].mean()
print(f"Average Profit Margin: {avg_margin:.2f}%")
```

```python
# total ad spend
total_ad_spend = df['Ad_Spend'].sum()
print(f"Total Ad Spend: ₹{total_ad_spend:,.2f}")
```

```python
# KPI(key performance indicator)
kpis = {
    'Total Revenue (₹)': f"{total_revenue:,.2f}",
    'Total Profit (₹)': f"{total_profit:,.2f}",
    'Units Sold': total_units,
    'Avg Customer Rating': f"{avg_rating:.2f}",
    'Avg Profit Margin (%)': f"{avg_margin:.2f}",
    'Total Ad Spend (₹)': f"{total_ad_spend:,.2f}"
}

pd.DataFrame.from_dict(kpis, orient='index', columns=["Value"])
```

## visualize your ecommerce data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# monthly Revenue Trend

```python
monthly_revenue = df.groupby('date')['Revenue'].sum()

plt.figure(figsize=(12, 6))
monthly_revenue.plot(marker='o')
plt.title('Monthly Revenue Trend')
plt.xlabel('Date')
plt.ylabel('Revenue (₹)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Revenue by Category

```python
category_revenue = df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=category_revenue.index, y=category_revenue.values)
plt.title('Revenue by Category')
plt.ylabel('Revenue (₹)')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Profit by Region

```python
region_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=region_profit.index, y=region_profit.values)
plt.title('Profit by Region')
plt.ylabel('Profit (₹)')
plt.xlabel('Region')
plt.tight_layout()
plt.show()
```

# Average Customer Rating by category

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x='Category', y='Customer_Rating',estimator='mean',ci = None, palette="viridis")
plt.title("Average Customer Rating by Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Ad spend vs Revenue(scatter plot)

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Ad_Spend', y='Revenue', hue='Category')
plt.title('Ad Spend vs Revenue')
plt.xlabel('Ad Spend (₹)')
plt.ylabel('Revenue (₹)')
plt.tight_layout()
plt.show()
```

# Category share

```python
df.groupby('Category')['Revenue'].sum().plot.pie(autopct='%1.1f%%', figsize=(6,6), title='Revenue Share by Category')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
